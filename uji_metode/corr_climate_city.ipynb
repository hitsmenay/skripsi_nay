```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Correlation between ISPA and PM 2.5 in DKI Jakarta
plt.figure(figsize=(10,5))
plt.title("Correlation of ISPA and PM 2.5")
cmap = sns.diverging_palette(260,-10,s=50, l=50, n=6, as_cmap=True) #pewarnaan

x = corr_pm25['ispa']
y = corr_pm25['pm2p5']
correl = np.corrcoef(x, y)

sns.heatmap(correl, square=True, annot=True, linewidths=.5, cmap=cmap, vmin=-1, vmax=1)
plt.show()
```

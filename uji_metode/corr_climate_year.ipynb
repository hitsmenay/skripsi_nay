```python
climate_2018 = df_climate[df_climate['year'] == 2018]
climate_2019 = df_climate[df_climate['year'] == 2019]
climate_2020 = df_climate[df_climate['year'] == 2020]
climate_2021 = df_climate[df_climate['year'] == 2021]
climate_2022 = df_climate[df_climate['year'] == 2022]
```

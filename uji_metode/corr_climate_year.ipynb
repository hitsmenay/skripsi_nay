```python
import pandas as pd
import xarray as xr
import numpy as np
```

```python
corr_pm25 = pd.read_csv("/Users/roniabuhanif/Downloads/BMKG/data_skripsi/uji_metode/allvar_ispa.csv")
corr_pm25 = corr_pm25.drop(['time'], axis=1)
```

```python
climate_2018 = corr_pm25[corr_pm25['year'] == 2018]
climate_2019 = corr_pm25[corr_pm25['year'] == 2019]
climate_2020 = corr_pm25[corr_pm25['year'] == 2020]
climate_2021 = corr_pm25[corr_pm25['year'] == 2021]
climate_2022 = corr_pm25[corr_pm25['year'] == 2022]
```

```python
#import modul
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
i_2018 = (climate_2018.columns[6:])
i_2019 = (climate_2019.columns[6:])
i_2020 = (climate_2020.columns[6:])
i_2021 = (climate_2021.columns[6:])
i_2022 = (climate_2022.columns[6:])
print(i_2018,i_2019,i_2020,i_2021,i_2022)
```
